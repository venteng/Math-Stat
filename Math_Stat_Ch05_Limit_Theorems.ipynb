# Mathematical Statistics Chapter 5 Limit Theorems

## 5.2 Low of large numbers

The figure of $f(x) = \frac{1}{2\pi} e^{-x^2/2}$ from 0 to 1.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x)
def f(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

# Generate x values from 0 to 1
x = np.linspace(0, 1, 400)

# Generate y values using the function
y = f(x)

# Plotting the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$f(x) = \frac{1}{2\pi} e^{-x^2/2}$")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the function f(x) = 1/(2π) * e^(-x²/2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.integrate import quad

# Function to compute the Riemann sum
def riemann_sum(f, a, b, N):
    # Width of each interval
    dx = (b - a) / N
    # x values at the left endpoint of each interval
    x = np.linspace(a, b-dx, N)
    # Summing up the function values at these points
    sum = np.sum(f(x) * dx)
    return sum

# Parameters for the Riemann sum
a = 0   # Lower limit
b = 1   # Upper limit
N = 10  # Number of intervals

# Calculate the Riemann sum
riemann_result = riemann_sum(f, a, b, N)

# Generate points for plotting
x_plot = np.linspace(a, b, N+1)
y_plot = f(x_plot)

# Plotting the function and the Riemann sum
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r"$f(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}$", zorder=1)
plt.bar(x_plot[:-1], y_plot[:-1], width=(b-a)/N, align='edge', edgecolor='red', alpha=0.3, label='Riemann Sum (Left Endpoint)', zorder=2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Riemann Sum Approximation of the Integral')
plt.legend()
plt.grid(True)
plt.show()

riemann_result
```
